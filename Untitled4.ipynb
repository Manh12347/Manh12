```python
import pandas as pd

# Đọc file CSV
file_path = '/content/drive/MyDrive/Py/subject-info.csv'  # Đảm bảo rằng bạn có đường dẫn chính xác tới file CSV
data = pd.read_csv(file_path)
đatatropna = data.dropna()
data.info()
```

# Mục mới

```python
datadropna = data.dropna()
datadropna.head(1000)
```

```python
# Sử dụng phương thức map để thay thế giá trị trong cột Gender
datadropna['Gender'] = datadropna['Gender'].map({'F': 1, 'M': 0})
datadropna.head(100)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ biểu đồ phân bố cho độ tuổi
plt.figure(figsize=(10, 6))

sns.histplot(datadropna['Age [years]'], kde=True, color='blue', bins=20)

plt.title('Age Distribution')
plt.xlabel('Age [years]')
plt.ylabel('Frequency')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Đếm số lượng Nam và Nữ
gender_counts = datadropna['Gender'].value_counts()

# Vẽ biểu đồ tròn
plt.figure(figsize=(8, 6))
plt.pie(gender_counts, labels=['Female', 'Male'], autopct='%1.1f%%', colors=['lightblue', 'lightcoral'], startangle=90)

# Tạo tiêu đề
plt.title('Gender Distribution')

# Hiển thị biểu đồ
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Vẽ biểu đồ phân bố cho mức đường huyết
plt.figure(figsize=(10, 6))

sns.histplot(datadropna['Glycaemia [mmol/l]'], kde=True, color='green', bins=20)

plt.title('Blood Glucose Distribution')
plt.xlabel('Blood Glucose [mmol/l]')
plt.ylabel('Frequency')

plt.show()
```

```python
# Loại bỏ cột "Blood pressure"
datadropna = datadropna.drop(columns=['Blood pressure [mmHg]'])

# Tính toán ma trận tương quan sau khi đã loại bỏ cột
correlation_matrix = datadropna.corr()

# Vẽ heatmap
plt.figure(figsize=(10, 8))  # Kích thước biểu đồ
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

```
